```python
# splitting data frame for access
X = DataFrame(data, columns=['production_budget_usd'])
Y = DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
# lets visualise the data now
# lets create scatter plot
matplotlib.pyplot.scatter(X,Y)
# show matplotlib
matplotlib.pyplot.show()
```

```python
# lets make chart prettier and code cleaner
# adding alias on the top to make it concise
plt.scatter(X,Y)
# show matplotlib
plt.show()
```

```python
# changing sixe of chart


plt.figure(figsize=(10,6))

# giving title to our chart
plt.title('Film cost vs GLobal revenue')
# making dots more transparent in scatter plot add aplha=0.3 in plt.scatter(X,Y)
plt.scatter(X,Y,alpha=0.3)
# adding labels to x and y axis
plt.xlabel('Production Budget $')
plt.ylabel('World Gross $')
# remove empty space below the graph, setting limit of 0 for least and max 300000000, set for y - ylim and for x - xlim
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
# show the graph
plt.show()
```

```python
# linear regression
```

```python
# importing linear regression in first cell
```

```python
# creating object for regression
regression = LinearRegression()
# run regression
regression.fit(X,Y)
```

```python
regression.coef_ 
```

```python
regression.intercept_
```

```python
# changing sixe of chart


plt.figure(figsize=(10,6))

# giving title to our chart
plt.title('Film cost vs GLobal revenue')
# making dots more transparent in scatter plot add aplha=0.3 in plt.scatter(X,Y)
plt.scatter(X,Y,alpha=0.3)
plt.plot(X,regression.predict(X),color='red',linewidth=4)
# adding labels to x and y axis
plt.xlabel('Production Budget $')
plt.ylabel('World Gross $')
# remove empty space below the graph, setting limit of 0 for least and max 300000000, set for y - ylim and for x - xlim
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
# show the graph
plt.show()
```

```python
# evaluate and analysis
```
